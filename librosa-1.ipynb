```python
import librosa
import numpy as np
import IPython.display as ipd
```

## Load audio file

```python
audio_data = 'audio-data/07ab324c602e4afab65ddbcc746c31b5.mp3'
x, sr = librosa.load(audio_data, offset=0, duration=26)
print(type(x), type(sr))
print(x.shape, sr)
```

## Play audio in notebook

```python
ipd.Audio(audio_data)
```

## Visualizing audio

```python
import matplotlib.pyplot as plt
import librosa.display 
plt.figure(figsize=(14,5))
librosa.display.waveshow(x, sr=sr)
plt.show()
```

### Spectrogram

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.title('Spectrogram')
plt.colorbar()
plt.show()
```

## Create a Audio Signal

```python
import soundfile as sf

sr = 22050
T = 5.0
t = np.linspace(0 ,T, int(T*sr))
x = 0.5*np.sin(2*np.pi*220*t)

ipd.Audio(x, rate=sr)
sf.write('tone_220.wav', x, sr)
```

```python
ipd.Audio('tone_220.wav')
```
